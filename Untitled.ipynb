```python
from matplotlib import pyplot as plt
from matplotlib import style 
style.use('ggplot')
```

```python
X = [12,14,15]
Y = [12,16,19]
plt.plot(X,Y,'g',label ='one',linewidth = 5)
plt.title('info')
plt.xlabel("X axis")
plt.ylabel("y_axis")
plt.legend()
plt.grid(True,color = 'k')
plt.show()
```
